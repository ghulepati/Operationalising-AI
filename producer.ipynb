```python
from kafka import KafkaProducer
import json
import time
import scipy.io as sio
import numpy as np

producer = KafkaProducer(
    bootstrap_servers=['localhost:9092'],
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# Load the .mat file
mat_data = sio.loadmat('/Users/saighule/Downloads/traffic_dataset.mat')
Y_train = mat_data['tra_Y_tr']

def send_message():
    for i in range(Y_train.shape[1]):  # Iterate over time steps
        message = {
            'timestamp': time.time(),
            'traffic_flow': Y_train[:, i].tolist()  # Convert numpy array to list
        }
        producer.send('traffic_flow_topic', message)
        print(f"Sent data point {i}")
        time.sleep(1)  # Simulate real-time data by waiting 1 second between messages

if __name__ == '__main__':
    send_message()
    producer.flush()
    producer.close()
```
